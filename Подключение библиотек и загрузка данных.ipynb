# 1. Создание/запись массива

```python
# Библиотека работает с массивами данных ( одномерными, двумерными, n-мерными)
```

```python
# Создадим массив данных
names = ['Саша','Маша','Света', 'Инга', 'Полина']
birth = [34, 78, 84, 96, 89]
ID = [1,2,3,4,5]
```

```python
# Объединим списки в один, используя функцию zip(). Можно подавать в качестве
# аргументов несколько списков через запятую
 print(list(zip(['Саша','Маша','Света', 'Инга', 'Полина'],birth)))
```

```python
babies = list(zip(names,birth))
babies
```

```python
# Переведем имеющийся набор данных(список) в объект DataFrame
# из библиотеки pandas
# под индексом (index) понимается нумерация строк,
# под колонками(columns) - названия стоблцов
df_1 = pd.DataFrame(data = babies, index = ID, columns = ['Имя', 'Количество родившихся'])
df_1
```
